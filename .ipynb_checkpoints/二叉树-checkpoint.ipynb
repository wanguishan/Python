```python
class TreeNode():
    """节点类"""
    def __init__(self, x):
        self.val = x
        self.left = None
        self.right = None
```

```python
class Tree():
    """树类"""
    def TreeDepth(self, pRoot):
        """二叉树的深度"""
        if pRoot:
            leftDepth = self.TreeDepth(pRoot.left) + 1
            rightDepth = self.TreeDepth(pRoot.right) + 1
            return max(leftDepth, rightDepth)
        else:
            return 0            

    def DLR(self, root):
        """递归实现先序遍历"""
        if not root:
            return
        print(root.val)
        self.DLR(root.left)
        self.DLR(root.right)
        # return [root.val] + self.DLR(root.left) + self.DLR(root.right)
        
    def LDR(self, root):
        """递归实现中序遍历"""
        if not root:
            return    
        self.LDR(root.left)
        print(root.val)
        self.LDR(root.right)
        
    def LDR(self, root):
        """递归实现后序遍历"""
        if not root:
            return    
        self.LDR(root.left)       
        self.LDR(root.right)
        print(root.val)
        
    def preorderTraversal(self, root):
        """利用栈实现树的先序遍历"""
        if not root:
            return []
        stack = [root]
        result = []
        while stack:
            node = stack.pop()
            result.append(node.val)
            if node.right:
                stack.append(node.right)
            if node.left:
                stack.append(node.left)
        return result
    
    def preorderTraversal(self, root):
        """利用栈实现树的先序遍历(二) """
        if not root:
            return []
        stack = []
        result = []
        while root or stack:
            if root:
                result.append(root.val)
                stack.append(root)
                root = root.left
            else:
                node = stack.pop()
                root = node.right
        return result    
    
    def inorderTraversal(self, root):
        """利用栈实现树的中序遍历"""     
        if not root:
            return []
        stack = []
        result = []
        while root or stack:
            if root:
                stack.append(root)
                root = root.left
            else:
                node = stack.pop()
                result.append(node.val)
                root = node.right
        return result
        
    def postorderTraversal(self, root):
        """利用栈实现树的后序遍历"""
        if not root:
            return []   
        stack = []
        result = []
        while root or stack:
            if root:
                result.append(root.val)
                stack.append(root)
                root = root.right
            else:
                node = stack.pop()
                root = node.left
        return result[::-1]

    def PrintFromTopToBottom(self, root):
        """利用队列实现树的层次遍历"""     
        if not root:
            return []
        queue = [root]
        result = []
        while queue:
            node = queue.pop(0)
            result.append(node.val)
            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)
        return result        
        
```
